```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from PIL import Image, ImageTk

class GenderRecognitionApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Gender Recognition App")

        self.file_path = tk.StringVar()
        self.gender_result = tk.StringVar()
        self.male_prob = tk.StringVar()
        self.female_prob = tk.StringVar()

        self.create_widgets()

    def create_widgets(self):
        # File Entry
        file_label = tk.Label(self.root, text="Select Audio File:")
        file_label.pack(pady=10)

        file_entry = tk.Entry(self.root, textvariable=self.file_path, width=40)
        file_entry.pack(pady=5, padx=10)

        browse_button = tk.Button(self.root, text="Browse", command=self.browse_file)
        browse_button.pack(pady=5)

        # Result Display
        result_label = tk.Label(self.root, text="Gender Recognition Result:")
        result_label.pack(pady=10)

        result_display = tk.Label(self.root, textvariable=self.gender_result, font=("Helvetica", 16, "bold"))
        result_display.pack()

        probability_display = tk.Label(self.root, textvariable=self.get_probability_text())
        probability_display.pack()

        # Predict Button
        predict_button = tk.Button(self.root, text="Predict Gender", command=self.predict_gender)
        predict_button.pack(pady=20)

    def browse_file(self):
        file_path = filedialog.askopenfilename(filetypes=[("Wave files", "*.wav")])
        if file_path:
            self.file_path.set(file_path)

    def predict_gender(self):
        file = self.file_path.get()

        if not file or not file.endswith(".wav"):
            messagebox.showerror("Error", "Please select a valid WAV file.")
            return

        # Call your gender recognition function here
        # ...

        # Mock result for demonstration purposes
        gender = "male"
        male_prob = 0.75
        female_prob = 0.25

        self.gender_result.set(f"Result: {gender.capitalize()}")
        self.male_prob.set(f"Male Probability: {male_prob * 100:.2f}%")
        self.female_prob.set(f"Female Probability: {female_prob * 100:.2f}%")

    def get_probability_text(self):
        return tk.StringVar(value="Probabilities:     Male: {male_prob*100:.2f}%    Female: {female_prob*100:.2f}%")


if __name__ == "__main__":
    root = tk.Tk()
    app = GenderRecognitionApp(root)
    root.mainloop()
```
